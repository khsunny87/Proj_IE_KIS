```python
def Get_data_from_log(f):
    fline=f.readline()

    if not fline:
        return None
    ret_list=list([fline])

    D_End=re.compile('(\\d{8})').findall(fline)[2]

    while(True):
        fline=f.readline()
        if not fline:
            break
        ret_list=ret_list+[fline]

        if D_End==re.compile('~(\\d{8})').findall(fline)[0]:
            break
    return ret_list
    
```
